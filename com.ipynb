```python
import matplotlib.pyplot as plt

def plot_comparison(aes_time, des_time, des3_time):
    algorithms = ['AES', 'DES', 'Triple DES']
    time = [aes_time, des_time, des3_time]

    # Plot the results
    plt.bar(algorithms, time,color='r')
    plt.ylabel('Time (seconds)')
    plt.title('Time Comparison for encryption of video')
    plt.show()

aes_time = float(input("Enter time taken for AES (in seconds): "))
des_time = float(input("Enter time taken for DES (in seconds): "))
des3_time = float(input("Enter time taken for Triple DES (in seconds): "))

plot_comparison(aes_time, des_time,des3_time)
```
